## Carregando as Bibliotecas e lendo os dados do arquivo CSV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:.2f}'.format

df = pd.read_csv('./marketing_campaign.csv', sep=';')
```

## Preparando e explorando os dados

```python
df.info()
```

### * Verificando dados ausentes

* Verificando se temos dados ausentes em nosso dataset.
* Removendo dados ausentes para ter uma análise mais acertiva.

```python
df.isnull().sum()
```

```python
df.Income = df.Income.fillna(0)
df.isnull().sum()
```

```python
df.info()
```

* Verificando valores e tipos para **educação** e **Status matrimonial**.
* Alterar tipo para **Categórico**.

```python
df['Education'].value_counts().to_frame()
```

```python
df['Marital_Status'].value_counts().to_frame()
```

```python
u = list(df['Marital_Status'].unique())
u.sort()
u
```

```python
df[['Education','Marital_Status']] = df[['Education','Marital_Status']].astype('category')
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df[['Education','Marital_Status','Dt_Customer']].info()
```

## Análise Estatistica Básica

```python
df.describe()
```

```python
df.describe(include='category')
# Pessoas Casadas e com Graduados são mais ativos
```

```python
pd.options.display.float_format = '{:.1f}%'.format
ed = df['Education'].value_counts(normalize=True)*100
ed
# Pessoas Graduadas ativas representam 50.3% se comparados com outros tipos de escolaridades
```

```python
df['Marital_Status'].value_counts(normalize=True)*100
# Pessoas Casadas Representam 38.6% dos Clientes Ativos
```

```python
pd.crosstab(df['Education'],
            df['Marital_Status'], normalize=True
            )*100

# Visulaização mais detalhadas dos Cliente ativos relacionados 
# por Status Civil e Educação
```

```python
plt.figure(figsize=(12,5))
sns.heatmap((pd.crosstab(df['Education'],
            df['Marital_Status'], normalize=True
            )*100).astype('int'), annot=True, fmt='d')
```
